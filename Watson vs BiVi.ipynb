# Cleaning dan Transformasi data

Cleaning data termasuk pembenaran format penanggalan, dan melakukan mapping code product <b> table baru bernama data </b>

## Menyamakan Format Penanggalan

<b> hal ini dilakukan karena format penanggalan pada januari 2021 data watson memiliki format yang berbeda sendiri

```python
def convert(tanggal):
    return datetime.datetime.strptime(tanggal[0],"%m/%d/%Y").strftime("%Y-%m-%d")

januari2021.Tanggal = januari2021.apply(convert,axis=1)
```
